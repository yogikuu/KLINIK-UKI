## TASK - PENJUALAN

### Rekap Penjualan

<p> Merekap penjualan perhari,minggu,bulan, kuartal dan tahun</p> 

```python
#import module date time 
from datetime import date
```

```python
#membuat kolom baru [day]
df_revenue_manipulation['day'] = df_revenue_manipulation['Transaction Date'].dt.day_of_year
#mengubah tipe data ke int
df_revenue_manipulation['day']=df_revenue_manipulation['day'].astype('int')

nilai_penjualan_day=df_revenue_manipulation.groupby(['day'])['Total Transaction'].sum()
display(nilai_penjualan_day.to_frame())



```

```python
df_revenue_manipulation['week'] = df_revenue_manipulation['Transaction Date'].dt.isocalendar().week
df_revenue_manipulation['week']=df_revenue_manipulation['week'].astype('int')

nilai_penjualan_minggu=df_revenue_manipulation.groupby(['week'])['Total Transaction'].sum()
display(nilai_penjualan_minggu.to_frame())
```

```python
df_revenue_manipulation['month'] = df_revenue_manipulation['Transaction Date'].dt.month
df_revenue_manipulation['month']=df_revenue_manipulation['month'].astype('int')
nilai_penjualan_bulan=df_revenue_manipulation.groupby(['month'])['Total Transaction'].sum()
display(nilai_penjualan_bulan.to_frame())
```

```python
df_revenue_manipulation['quarter'] = df_revenue_manipulation['Transaction Date'].dt.quarter
df_revenue_manipulation['quarter']=df_revenue_manipulation['quarter'].astype('int')
nilai_penjualan_kuartal=df_revenue_manipulation.groupby(['quarter'])['Total Transaction'].sum()
display(nilai_penjualan_kuartal.to_frame())
```

```python
df_revenue_manipulation['year'] = df_revenue_manipulation['Transaction Date'].dt.year
df_revenue_manipulation['year']=df_revenue_manipulation['year'].astype('int')
nilai_penjualan_tahun=df_revenue_manipulation.groupby(['year'])['Total Transaction'].sum()
display(nilai_penjualan_tahun.to_frame())
```

### Rekap Penjualan per produk

```python
df_revenue_manipulation.head()
```

```python

#melakukan grouping terlebih dahulu berdasarkan kolom item dan menambahkan fungsi agregasi 'sum' terhadap 'total transaction'
df_rekap_penjualan_produk=df_revenue_manipulation.groupby(['Item'])['Total Transaction'].sum()
print('nilai produk yang terjual')
#data frame yang sudah di gruping kemudian di sorting secara descending berdasarkan 'total transaction'
#kemudian ditampilkan berdasarkan nilai transaction
display(df_rekap_penjualan_produk.to_frame().sort_values('Total Transaction',ascending=False))
print()
print()

#melihat jumlah produk yang terjual
produk_terjual=df_revenue_manipulation.value_counts('Item')
print('jumlah produk yang terjual')
display(produk_terjual)
print()
print()

#melihat persentase jumlah produk yang terjual
print('persentase produk yang terjual')
display(df_revenue_manipulation.value_counts('Item',normalize=True))
```

### Rekap jumlah jenis Customer

ada dua jenis customer di pada customer yaitu customer 'walk in' dan 'instansi', diasumsikan sebagai berikut:

- customer walkin yaitu customer yang payernya adalah 'UMUM'
- customer instansi yaitu custome yang payernya selain 'UMUM'

```python
df_revenue_manipulation.head()
```

```python
import numpy as np

df_revenue_manipulation['jenis_customer'] = np.where((df_revenue_manipulation['Payer']=='UMUM'), 'Walk in', 'instansi')
df_jenis_customer=df_revenue_manipulation[[ 'Registration No', 'jenis_customer','month']]
df_jenis_customer=df_jenis_customer.groupby(['month','Registration No','jenis_customer'],as_index=False).count()
display(df_jenis_customer)
df_jenis_customer.pivot_table(values='jenis_customer', index='month', columns='jenis_customer', aggfunc='count')
```

### Rekap Penjualan Payment status

```python
# rekap penjualan tunai,kredit, maupun in care

penjualan_payment=df_revenue_manipulation.groupby('Payment Status')['Total Transaction'].sum()

display(penjualan_payment.to_frame())

```

### Rekap pasien BPJS per bulan

```python
df_temp=df_revenue_manipulation[['Registration No','Payer','month',]]

df_payer_BPJS=df_temp[df_temp['Payer']=='BPJS']
df_payer_BPJS=df_payer_BPJS.groupby(['month','Registration No','Payer'], as_index=False).count()
df_payer_BPJS
df_payer_BPJS=df_payer_BPJS.pivot_table(values='Payer', index='Payer', columns='month', aggfunc='count')
df_payer_BPJS
#df_payer_BPJS.groupby('month')['Payer'].count()
```

## Rekap nilai pengadaan

```python
#membuat kolom baru [day]
df_pengadaan['day'] = df_pengadaan['Tgl Terima'].dt.day_of_year
#mengubah tipe data ke int
df_pengadaan['day']=df_pengadaan['day'].astype('int')

nilai_pengadaan_day=df_pengadaan.groupby(['day'])['Jumlah Harga'].sum()
display(nilai_pengadaan_day.to_frame())
```

```python
#membuat kolom baru [week]
df_pengadaan['week'] = df_pengadaan['Tgl Terima'].dt.isocalendar().week
#mengubah tipe data ke int
df_pengadaan['week']=df_pengadaan['week'].astype('int')

nilai_pengadaan_week=df_pengadaan.groupby(['week'])['Jumlah Harga'].sum()
display(nilai_pengadaan_week.to_frame())
```

```python
#membuat kolom baru [month]
df_pengadaan['month'] = df_pengadaan['Tgl Terima'].dt.month
#mengubah tipe data ke int
df_pengadaan['month']=df_pengadaan['month'].astype('int')

nilai_pengadaan_month=df_pengadaan.groupby(['month'])['Jumlah Harga'].sum()
display(nilai_pengadaan_month.to_frame())
```

```python
#membuat kolom baru [quarter]
df_pengadaan['quarter'] = df_pengadaan['Tgl Terima'].dt.quarter
#mengubah tipe data ke int
df_pengadaan['quarter']=df_pengadaan['quarter'].astype('int')

nilai_pengadaan_quarter=df_pengadaan.groupby(['quarter'])['Jumlah Harga'].sum()
display(nilai_pengadaan_quarter.to_frame())
```

```python
#membuat kolom baru [quarter]
df_pengadaan['year'] = df_pengadaan['Tgl Terima'].dt.year
#mengubah tipe data ke int
df_pengadaan['year']=df_pengadaan['year'].astype('int')

nilai_pengadaan_year=df_pengadaan.groupby(['year'])['Jumlah Harga'].sum()
display(nilai_pengadaan_year.to_frame())
```
